```python
user = open("key.txt",'r').read()
key = open("key.txt",'r').read().split()
service = webdriver.chrome.service.Service('./chromedriver')
service.start()
options = webdriver.ChromeOptions()
#options.add_argument('--headless')
options = options.to_capabilities()
driver = webdriver.Remote(service.service_url, options)
driver.get('https://www.qrcrypher.com/login')
driver.implicitly_wait(10)
username = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/form/div/div[1]/div/input')
password = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/form/div/div[2]/div/input')
username.send_keys(key[0])
password.send_keys(key[1])
driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/form/div/div[4]/div/div/button').click()
```

```python
starttime=time.time()
timeout = time.time() + 60*60*20  # 60 seconds times 360 meaning 6 hrs
while time.time() <= timeout: 
    try:
        request_token1 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[1]/div/p[1]').text
        name1 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[1]/div/input')
        name1.send_keys(request_token1)
        request_token2 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[1]/div/p[2]').text
        name2 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[3]/div/input')
        name2.send_keys(request_token2)
        request_token3 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[1]/div/p[3]').text
        name3 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[5]/div/input')
        name3.send_keys(request_token3)
        request_token4 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[1]/div/p[4]').text
        name4 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[7]/div/input')
        name4.send_keys(request_token4)
        request_token5 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[1]/div/p[5]').text
        name5 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[9]/div/input')
        name5.send_keys(request_token5)
        request_token6 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[2]/div/p[1]').text
        name6 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[2]/div/input')
        name6.send_keys(request_token6)
        request_token7 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[2]/div/p[2]').text
        name7 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[4]/div/input')
        name7.send_keys(request_token7)
        request_token8 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[2]/div/p[3]').text
        name8 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[6]/div/input')
        name8.send_keys(request_token8)
        request_token9 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[2]/div/p[4]').text
        name9 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[8]/div/input')
        name9.send_keys(request_token9)
        request_token10 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[2]/div/div/div[2]/div/p[5]').text
        name10 = driver.find_element_by_xpath('/html/body/section[2]/div/div/div[3]/div/form/div/div[10]/div/input')
        name10.send_keys(request_token10)
        driver.find_element_by_xpath('/html/body/section[2]/div/div/div[4]/div/button').click()
        driver.implicitly_wait(20)
        driver.find_element_by_xpath('/html/body/div/div/div[3]/button[1]').click()
        time.sleep(3 - ((time.time() - starttime) % 3.0))
    except KeyboardInterrupt:
        print('\n\nKeyboard exception received. Exiting.')
        exit()
```
